```python
def get_whu_siri_labels(labels):
    text_labels = ['农业', '商业', '港口', '闲置土地', '工业',
                   '草甸', '立交桥', '公园', '池塘', '住宅', '河', '水']
    return [text_labels[int(i)] for i in labels]
def get_UC_labels(labels):
    text_labels = ['农业','飞机','棒球场','海滩','建筑物','丛林','密集住宅','森林','高速公路','高尔夫球场'
,'港口','路口','中型住宅','移动房公园','立交桥','停车场','河','跑道','稀疏住宅','储油罐','网球场']
    return [text_labels[int(i)] for i in labels]
```

```python
def show_whu_siri(images, labels):
    d2l.use_svg_display()
    # 这里的_表示我们忽略（不使用）的变量
    _, axes = plt.subplots(nrows = 2, ncols = 6)
#     for f, img, lbl in zip(figs, images, labels):
    for i in range(2):
        for j in range(6):
            axes[i][j].imshow(images[i*6+j])
            axes[i][j].set_title(labels[i*6+j], fontproperties = 'SimHei')
            axes[i][j].axes.get_xaxis().set_visible(False)
            axes[i][j].axes.get_yaxis().set_visible(False)
    plt.show()
def show_UC(images, labels):
    d2l.use_svg_display()
    # 这里的_表示我们忽略（不使用）的变量
    _, axes = plt.subplots(nrows = 3, ncols = 7)
#     for f, img, lbl in zip(figs, images, labels):
    for i in range(3):
        for j in range(7):
            axes[i][j].imshow(images[i*7+j])
            axes[i][j].set_title(labels[i*7+j], fontproperties = 'SimHei')
            axes[i][j].axes.get_xaxis().set_visible(False)
            axes[i][j].axes.get_yaxis().set_visible(False)
    plt.show()
```

```python
import matplotlib.pyplot as plt
X, y = [], []
for i in range(5,420,20):
    X.append(dataset[i][0])
    y.append(dataset[i][1])
show_UC(X, get_UC_labels(y))
```
